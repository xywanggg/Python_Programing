```python
class Vehicle:
    def __init__(self, vemk, vemd, veyr, veml):
        self.__vehicle_make = vemk
        self.__vehicle_model = vemd
        self.__vehicle_year = veyr
        self.__vehicle_mileage = veml
        
    def get_make(self):
        return self.__vehicle_make
    def get_made(self):
        return self.__vehicle_model
    def get_year(self):
        return self.__vehicle_year
    def get_mileage(self):
        return self.__vehicle_mileage
    
    def set_make(self,vemk):
        self.__vehicle_make = vemk
    def set_made(self,vemd):
        self.__vehicle_model = vemd
    def set_year(self,veyr):
        self.__vehicle_year = veyr
    def set_mileage(self,veml):
        self.__vehicle_mileage = veml
        
    def __str__(self):
        return 'Vehicle make: '+self.__vehicle_make+'; Vehicle model: '+self.__vehicle_model+'; Vehicle year of manufacture:'+str(self.__vehicle_year)+'; Vechicle mileage:'+str(self.__vehicle_mileage)
```

```python
#from employee import Employee
class FullTimeEmployee(Employee):
    def __init__(self,emnm, emadd,ve_mk, ve_md, ve_yr, ve_ml, sal):
        super().__init__(emnm, emadd, ve_mk, ve_md, ve_yr, ve_ml)
        self.__salary = sal
        
    def get_salary(self):
        return self.__salary
    
    def set_salary(self,sal):
        self.__salary = sal
        
    def compute_compensation(self):
        sa = float(self.__salary)
        if sa <= 45000:
            return sa-sa*0.18
        elif sa>45000 and sa<=82000:
            return sa-45000*0.18-(sa-45000)*0.28
        else:
            return sa-45000*0.18-(82000-45000)*0.28-(sa-82000)*0.33

    def __str__(self):
        childData = super().__str__()
        childData = '\nDetails of the full time employee is:'+childData
        childData += '\nThe salary is {}'.format(self.__salary)
        return childData
    
```

```python
#from employee import Employee
class PartTimeEmployee(Employee):
    def __init__(self,emnm, emadd, ve_mk, ve_md, ve_yr, ve_ml, hr, rate ):
        super().__init__(emnm, emadd, ve_mk, ve_md, ve_yr, ve_ml)
        self.__hoursWorked = hr
        self.__hourlyRate = rate
        
    def get_hoursWorked(self):
        return self.__hoursWorked
    def get_hourlyRate(self):
        return self.__hourlyRate
    
    def set_hoursWorked(self, hr):
        self.__hoursWorked = hr
    def set_hourlyRate(self, rate):
        self.__hourlyRate = rate
        
    def compute_compensation(self):
        h = int(self.__hoursWorked)
        r = int(self.__hourlyRate)
        if h <40:
            return h*r
        else:
            return 40*r+(h-40)*r*1.8
            
    def __str__(self):
        childData = super().__str__()
        childData = '\nDetails of the part time employee is:'+childData
        childData += '\nThe hourly rate is {} \nthe number of hours worked is {}'.format(self.__hourlyRate, self.__hoursWorked)
        return childData
```
